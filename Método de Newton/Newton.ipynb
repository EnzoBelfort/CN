```python
import matplotlib.pyplot as plt
import numpy as np


def funcao1(x):
    return x * np.exp(-x)


def funcao2(x):
    return x**3 - x - 3


def funcao3(x):
    return np.arctan(x)


x = np.linspace(-5, 5, 400)


y1 = funcao1(x)
y2 = funcao2(x)
y3 = funcao3(x)


plt.figure(figsize=(6, 4))
plt.plot(x, y1, label='f(x) = x * e^(-x)', color='blue')
plt.title('Gráfico de f(x) = x * e^(-x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(6, 4))
plt.plot(x, y2, label='f(x) = x^3 - x - 3', color='green')
plt.title('Gráfico de f(x) = x^3 - x - 3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(6, 4))
plt.plot(x, y3, label='f(x) = arctan(x)', color='red')
plt.title('Gráfico de f(x) = arctan(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math

def metodo_newton(f, f_derivada, x0, epsilon, maxit):
    print("k\t xk\t\t f(xk)\t\t f'(xk)\t\t step")
    print("-" * 50)

    xk = x0
    fk = f(xk)
    fpk = f_derivada(xk)
    sk = float('inf')  # Passo inicial
    k = 0

    while abs(fk) > epsilon and abs(sk) > epsilon and k < maxit:
        if abs(fpk) < epsilon:
            print("Derivada muito próxima de zero. Escolha outro valor inicial.")
            return None, None, k

        xk1 = xk - fk / fpk  # Atualização pelo Método de Newton
        sk = xk1 - xk        # Tamanho do passo

        # Atualiza valores
        xk = xk1
        fk = f(xk)
        fpk = f_derivada(xk)

        # Imprime resultados da iteração
        print(f"{k}\t {xk:.6f}\t {fk:.6e}\t {fpk:.6e}\t {sk:.6e}")

        k += 1

    print("-" * 50)

    if abs(fk) <= epsilon:
        print("Convergência atingida: |f(xk)| < epsilon")
    elif abs(sk) <= epsilon:
        print("Convergência atingida: |sk| < epsilon")
    else:
        print("Número máximo de iterações atingido.")

    return xk, fk, k

# Exemplo de uso:
def f(x):
    return x * math.exp(-x)

def f_derivada(x):
    return math.exp(-x) - x * math.exp(-x)

x0 = 2.0
epsilon = 1e-8
maxit = 20

x, fx, iteracoes = metodo_newton(f, f_derivada, x0, epsilon, maxit)

print(f"\nResultado final:\nSolução aproximada: x = {x:.6f}\nValor de f(x): {fx:.6e}\nNúmero de iterações: {iteracoes}")
```
